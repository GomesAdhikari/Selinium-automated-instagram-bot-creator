```python
import pandas as pd
import concurrent.futures
import requests

# Load the proxy DataFrame
# combined_df = pd.read_csv('all_proxy.csv')  # Make sure combined_df is already loaded with your proxy data

# Function to check if a proxy is valid
def check_proxy(proxy):
    try:
        # Define the proxy dictionary
        proxies = {
            "http": f"http://{proxy}",
            "https": f"http://{proxy}",
        }
        response = requests.get("http://instagram.com", proxies=proxies, timeout=5)
        response_2 = requests.get('https://temp-mail.io/en',proxies=proxies, timeout=5)
        if (response.status_code == 200) and (response_2.status_code == 200):
            return proxy  # Return the proxy if it's valid
    except requests.exceptions.RequestException:
        return None  # Return None if the proxy fails
df = pd.read_csv('all_proxy.csv')
# Use ThreadPoolExecutor for concurrent proxy checking
with concurrent.futures.ThreadPoolExecutor(max_workers=50) as executor:  # Increase max_workers as needed
    futures = [executor.submit(check_proxy, proxy) for proxy in df['proxy']]
    valid_proxies = [future.result() for future in concurrent.futures.as_completed(futures) if future.result() is not None]

# Print the valid proxies
print("Valid proxies:", valid_proxies)
with open('working_proxies.txt', 'w') as file:
    for proxy in valid_proxies:
        file.write(proxy + '\n')

print("Valid proxies have been saved to 'valid_proxies.txt'")
```
